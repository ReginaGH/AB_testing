### Расчёт размера выборки и длительности теста
Для эксперимента будем использовать 20% от суточного трафика: 50_000 * 0,2 со сплитованием 50/50 с улучшением конверсии от 0,7 до 0,9

```python
# Вводные данные:
nobs_1, nobs_2 = 5_000, 5_000
conv_1, conv_2 = 0.7, 0.9
prob_1, prob_2 = conv_1/nobs_1, conv_2/nobs_2
alpha = 0.05
beta = 0.2
```

```python
# Рассчитаем нормированный эффект:
effect_size = proportion_effectsize(prob_1, prob_2)
# Рассчитаем размер выборки на обе группы:
n = zt_ind_solve_power(effect_size=effect_size, alpha=alpha, power=1-beta, ratio = 1) * 2
# Рассчитаем кол-во дней, котрые необходимы для проведния эксперимента:
days = n/(nobs_1+nobs_2)
print(f'Sample size for both tests groups is {np.round(n, 0)}, considering effect size equals {np.round(effect_size, 5)}\n Tests duration is {days} days')
```

**Вывод**: ~ 10 мес длительность теста - не целесообразно. Это на тех данных, которые получены через поиск, возможно на реальных данных будут другие результаты.
